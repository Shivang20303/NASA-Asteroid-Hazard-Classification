```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = Path.cwd().parent
Path = base / "input_data" / "nasa.csv"
# print(Path)
```

```python
df = pd.read_csv(Path)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# Dropping ID's, dates and other columns that wont help in prediction
df = df.drop(["Neo Reference ID", "Name", "Orbit ID", "Close Approach Date", "Epoch Date Close Approach", "Orbit Determination Date"], axis=1)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encode = LabelEncoder()
df["Hazardous"] = encode.fit_transform(df["Hazardous"])

df.head()
```

```python
df["Orbiting Body"].value_counts()
```

```python
df["Equinox"].value_counts()
```

```python
# Dropping them since they have only one value (so won't help prediction)
df = df.drop(["Orbiting Body", "Equinox"], axis=1)
df.head()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
df = df.drop([
    "Est Dia in KM(max)",
    "Est Dia in M(min)",
    "Est Dia in M(max)",
    "Est Dia in Miles(min)",
    "Est Dia in Miles(max)",
    "Est Dia in Feet(max)",
    "Relative Velocity km per hr",
    "Miles per hour",
    "Miss Dist.(lunar)",
    "Miss Dist.(kilometers)",
    "Miss Dist.(miles)"
], axis=1)
```

```python
df.info()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
df.describe()
```

```python
X = df.drop("Hazardous", axis=1)
y = df["Hazardous"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=101)
```

```python
from xgboost import XGBClassifier, plot_importance
from matplotlib import pyplot
```

```python
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)
plot_importance(xgb_model)
pyplot.show()
```

```python
from lightgbm import LGBMClassifier
```

```python
light_model = LGBMClassifier()
light_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
y_pred_xgb = xgb_model.predict(X_test) 
y_pred_light = light_model.predict(X_test) 
acc = accuracy_score(y_test, y_pred_xgb)*100
print("XGBoost Model")
print(f"Accuracy: {acc}%")
print(f"Classification Report: \n{classification_report(y_test, y_pred_xgb)}")
print(f"Confusion Matrix: \n{confusion_matrix(y_test, y_pred_xgb)}")

print("\n\nLightGBM Model")
print(f"Accuracy: {acc}%")
print(f"Classification Report: \n{classification_report(y_test, y_pred_light)}")
print(f"Confusion Matrix: \n{confusion_matrix(y_test, y_pred_light)}")
```

## Testing with parameters

```python
xgb_model = XGBClassifier(n_estimators=100, learning_rate=0.1, eval_metric="logloss", random_state=101)
xgb_model.fit(X_train, y_train)
plot_importance(xgb_model)
pyplot.show()
```

```python
light_model = LGBMClassifier(n_estimators=100, learning_rate=0.1, random_state=101)
light_model.fit(X_train, y_train)
```

```python
y_pred_xgb = xgb_model.predict(X_test) 
y_pred_light = light_model.predict(X_test) 
acc = accuracy_score(y_test, y_pred_xgb)*100
print("XGBoost Model")
print(f"Accuracy: {acc}%")
print(f"Classification Report: \n{classification_report(y_test, y_pred_xgb)}")
print(f"Confusion Matrix: \n{confusion_matrix(y_test, y_pred_xgb)}")

print("\n\nLightGBM Model")
print(f"Accuracy: {acc}%")
print(f"Classification Report: \n{classification_report(y_test, y_pred_light)}")
print(f"Confusion Matrix: \n{confusion_matrix(y_test, y_pred_light)}")
```
